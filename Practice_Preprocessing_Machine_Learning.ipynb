Fillna (Solution 2)

```python
dfFill = df.copy()
dfFillZero = df.copy()
```

```python
print (dfFill.isnull().sum())
```

```python
numeric_cols = dfFill.select_dtypes(include=['number']).columns
dfFill[numeric_cols] = dfFill[numeric_cols].fillna(dfFill[numeric_cols].mean())

print (dfFill.isnull().sum())
dfFill
```

Scikit-learn (Solution 3)

```python
X = df[['Country', 'Age', 'Salary']].values
y = df['Purchased'].values
```

Mean

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
print(X)
```

Median

```python
X = df[['Country', 'Age', 'Salary']].values
y = df['Purchased'].values
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
print(X)
```

Most_frequent

```python
X = df[['Country', 'Age', 'Salary']].values
y = df['Purchased'].values
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
print(X)
```

Constant

```python
X = df[['Country', 'Age', 'Salary']].values
y = df['Purchased'].values
imputer = SimpleImputer(missing_values=np.nan, strategy='constant')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
print(X)
```
